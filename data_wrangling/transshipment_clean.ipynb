```python
#other approach to get the names from latitude and longitude

import geopandas as gpd

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
gdf = gpd.GeoDataFrame(ts, geometry=gpd.points_from_xy(ts.mean_latitude, ts.mean_longitude))
```

```python
ts_df = gpd.sjoin(gdf, world, how='left')
```

```python
#check if it worked

ts_df.head()
```

```python
#check if country names make sense

ts_df['name'].value_counts()
```

```python
#droping columns not needed

drop_cols = ['geometry','index_right','pop_est','continent','gdp_md_est']
ts_df = ts_df.drop(drop_cols, axis=1)
```

```python
#confirm
ts_df.head()
```

```python
#check for nulls. not dropping them yet

ts_df['name'].isnull().value_counts()
```

```python
#rename columns

ts_df.columns=['fishing_vessel_mmsi', 'transshipment_vessel_mmsi', 'start_time',
       'end_time', 'mean_latitude', 'mean_longitude', 'duration_hr',
       'median_distance_km', 'median_speed_knots', 'coordinates', 'country',
       'iso_3']
```

```python
#confirm

ts_df.head()
```

```python
#check the types

ts_df.dtypes
```

```python
#change to date time

ts_df['start_time']=pd.to_datetime(ts_df['start_time'],errors='coerce')

ts_df['end_time']=pd.to_datetime(ts_df['end_time'],errors='coerce')
```

```python
#check the years to keep

ts_df['start_time'].dt.year.value_counts()
```

```python
#check the table

ts_df.head()
```

```python
#import as CSV

ts_df.to_csv('ts_clean.csv', index=False)
```
